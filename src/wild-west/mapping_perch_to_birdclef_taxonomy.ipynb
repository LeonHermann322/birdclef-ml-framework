```python
perch_taxonomy = pd.read_csv(yaml_config.base_data_dir + "/perch_taxonomy_labels.csv").rename(columns={"inat2024_fsd50k":"scientific_name"})
birdclef_taxonomy = pd.read_csv(yaml_config.base_data_dir + "/taxonomy.csv")
```

```python
# Calculate exact scientific name matches
exact_matches = set(perch_taxonomy["scientific_name"]).intersection(set(birdclef_taxonomy["scientific_name"]))
print(f"Number of exact matches: {len(exact_matches)}")
```

```python
missing_mappings = set(birdclef_taxonomy["scientific_name"]).difference(set(perch_taxonomy["scientific_name"]))
print(f"Number of missing mappings: {len(missing_mappings)}")

# Write missing mappings to csv
missing_mappings_df = birdclef_taxonomy[birdclef_taxonomy["scientific_name"].isin(missing_mappings)]
missing_mappings_df.to_csv("missing_mappings.csv", index=False)
```

```python
remaining_perch_taxonomy = perch_taxonomy[~perch_taxonomy["scientific_name"].isin(exact_matches)]

# Write remaining perch taxonomy to csv
remaining_perch_taxonomy.to_csv("remaining_perch_taxonomy.csv", index=False)
```
