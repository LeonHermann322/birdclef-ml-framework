```python
# Read and show label file
import pandas as pd

file = "/sc/scratch/leon.hermann/kaggle/competitions/birdclef-2026/train.csv"
labels = pd.read_csv(file)
labels.head()
```

```python
from src.datasets.birdclef_dataset import LabelEncoder, BirdClefSample

label_encoder = LabelEncoder(taxonomy_file="/sc/scratch/leon.hermann/kaggle/competitions/birdclef-2026/taxonomy.csv")

# Read each row and create a BirdClefSample
samples = []
for idx, (row_index, row) in enumerate(labels.iterrows()):
    if idx >= 5:
        break
    sample = BirdClefSample.from_audio_label(row, label_encoder, ds_path="/sc/scratch/leon.hermann/kaggle/competitions/birdclef-2026/train_audio")
    samples.append(sample)
```
